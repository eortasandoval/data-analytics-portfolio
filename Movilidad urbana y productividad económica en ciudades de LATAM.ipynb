### 2.3 Corregir formatos numéricos y de fecha

**🎯Objetivo:**
Asegurar que las columnas de fechas y valores numéricos estén en formatos correctos para permitir análisis, cálculos y comparaciones precisas.

**Instrucciones:**

- Convierte las columnas de fecha de `traffic` a formato `datetime`. Haz el cambio a prueba de errores.
- En el dataset `eco`, limpia los valores numéricos:
    - En `city_gdp_capita`: elimina separadores de miles (`.`) y reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
    - En `unemployment_pct`: elimina el símbolo de porcentaje (`%`) y reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
    - En `population_m`: reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
- Finalmente, crea una nueva columna llamada `population` multiplicando `population_m` por 1,000,000 para obtener la población total.


<details>
<summary>Haz clic para ver la pista</summary>
para eliminar símbolos, puedes reemplazarlos por un texto vacío.

```python

# Convertir las columnas de traffic a tipo fecha con pd.to_datetime()
traffic['update_time_utc'] = pd.to_datetime(
    traffic['update_time_utc'], errors='coerce'
)

traffic['update_time_utc_week_ago'] = pd.to_datetime(
    traffic['update_time_utc_week_ago'], errors='coerce'
)

# verificar el cambio
traffic.info()
```

```python
# Limpia separadores y convierte columnas numéricas en eco

# =========================
# Limpiar y convertir columnas numéricas en eco
# =========================

# city_gdp_capita: quitar separadores de miles y convertir a float
eco['city_gdp_capita'] = (
    eco['city_gdp_capita']
    .astype(str)
    .str.replace('.', '', regex=False)
    .str.replace(',', '.', regex=False)
    .astype(float)
)

# unemployment_pct: quitar % y convertir a float
eco['unemployment_pct'] = (
    eco['unemployment_pct']
    .astype(str)
    .str.replace('%', '', regex=False)
    .str.replace(',', '.', regex=False)
    .astype(float)
)

# population_m: convertir a float
eco['population_m'] = (
    eco['population_m']
    .astype(str)
    .str.replace(',', '.', regex=False)
    .astype(float)
)

# =========================
# Calcular población total
# =========================
eco['population'] = eco['population_m'] * 1_000_000

# verificar cambio
eco.info()
eco.head(3)
```

## 🧩Paso 4: Analizar y resumir datos de movilidad

Como el dataset de tráfico contiene **múltiples registros por ciudad**. En esta parte, calcularás los promedios anuales por ciudad para simplificar el análisis y obtener una visión más clara de las tendencias generales.

### 4.1 Calcular promedios de tráfico por ciudad

**🎯Objetivo:**
Obtener una vista consolidada del tráfico promedio por ciudad y año, para analizar patrones generales sin depender de datos diarios.

**Instrucciones**

- Agrupa los datos por `city`, `country` y `year`.
- Calcula el promedio **solo de las métricas de tráfico más relevantes**: como `jams_delay`, `traffic_index_live`, `jams_length_kms`, `jams_count`, `mins_delay`, y tiempos de viaje (`travel_time_live_per_10kms_mins` y `travel_time_hist_per_10kms_mins`).
- Guarda el resultado como `traffic_city_year_2024`, mantén las columnas como variables (no índices).


<details>
<summary>Haz clic para ver la pista</summary>
Usa ".agg()" para aplicar funciones de promedio. Al final, reinicia el índice para mantener las columnas de la agrupación como variables (no índices).

```python
# Calcular los  promedios de trafico por ciudad, país y año
traffic_city_year_2024 = (
    traffic_2024
    .groupby(['city', 'country', 'year'], as_index=False)[
        [
            'jams_delay',
            'traffic_index_live',
            'jams_length_kms',
            'jams_count',
            'mins_delay',
            'travel_time_live_per_10kms_min',
            'travel_time_historic_per_10kms_min'
        ]
    ]
    .mean()
)

# Mostrar resultado
traffic_city_year_2024.head()
```

## 🧩Paso 5: Unir movilidad y economía

Combinar datasets te permite analizar cómo se relacionan los indicadores económicos con los de movilidad.

### 5.1 Unir tráfico (tabla principal) con indicadores económicos

**🎯Objetivo:**
Combinar la información de tráfico y economía en un solo DataFrame para analizar cómo las condiciones económicas se relacionan con la movilidad urbana.

**Instrucciones**
- Selecciona solo las **columnas relevantes** de cada dataset (por ejemplo, variables clave de tráfico y de economía).
- Usa `.copy()` al crear subconjuntos para evitar modificar el dataset original.
- Une ambos DataFrames y define como **claves de unión** a `city` y `year`.
- Mantén solo las ciudades y años presentes en ambos datasets.
- Guarda el resultado en una nueva variable llamada `merged` y muestra las primeras 5 filas.


<details>
<summary>Haz clic para ver la pista</summary>
Aplica una unión de tipo "inner" para mantener las ciudades y años presentes en ambos datasets.

```python

# Seleccionar columnas clave de tráfico
left_cols = [
    'city',
    'country',
    'year',
    'jams_delay',
    'traffic_index_live',
    'jams_length_kms',
    'jams_count',
    'mins_delay',
    'travel_time_live_per_10kms_min',
    'travel_time_historic_per_10kms_min'
]

# Seleccionar columnas clave de economía (USANDO NOMBRES REALES)
right_cols = [
    'city',
    'year',
    'city_gdp_capita',
    'unemployment_pct',
    'pm2.5_μg_m3',
    'population'
]

# Crear datasets reducidos con copy()
traffic_2024_small = traffic_city_year_2024[left_cols].copy()
eco_2024_small = eco_2024[right_cols].copy()

# Unir datasets (inner join)
merged = pd.merge(
    traffic_2024_small,
    eco_2024_small,
    on=['city', 'year'],
    how='inner'
)

# Mostrar las primeras 5 filas
merged.head()


```

**Contexto & objetivo:**  
El objetivo de este análisis es explorar la relación entre la movilidad urbana y la productividad económica en ciudades de América Latina, evaluando si mayores niveles de congestión vehicular y tiempos de viaje más largos están asociados con menores niveles de PIB per cápita. Desde una perspectiva de política pública y planeación urbana, esta relación es clave para identificar ciudades donde la movilidad podría estar actuando como un cuello de botella estructural para el crecimiento económico.

Para ello, se utilizaron variables de movilidad como JamsDelay (minutos de retraso por congestión), Traffic Index Live y tiempos de viaje por cada 10 km, las cuales reflejan la eficiencia del sistema vial urbano. Estas métricas se contrastaron con indicadores económicos como PIB per cápita, tasa de desempleo y población, que permiten aproximar el nivel de productividad y bienestar económico de cada ciudad. La combinación de estas variables resulta fundamental para la toma de decisiones de inversión en infraestructura de transporte, al permitir priorizar ciudades donde mejorar la movilidad puede generar mayores impactos económicos y sociales.

**Cobertura de datos:**  
El análisis se realizó utilizando información correspondiente al año 2024, seleccionando este periodo por ser el más reciente y relevante para la evaluación de condiciones actuales de movilidad urbana y desempeño económico. Tras los procesos de limpieza, estandarización y filtrado, el dataset final incluye un conjunto de ciudades de América Latina presentes simultáneamente en ambas fuentes de datos (movilidad y economía).

La base consolidada integra información de múltiples ciudades latinoamericanas distribuidas en varios países de la región, incluyendo economías representativas como México, Brasil, Argentina, Colombia, Perú, Chile y Uruguay, entre otros. Cada observación del dataset final corresponde a una ciudad–año, lo que permite comparaciones homogéneas entre ciudades bajo un mismo marco temporal y facilita el análisis de patrones regionales entre congestión vehicular y productividad económica.

**Metodología (alto nivel):**  
El análisis se desarrolló a partir de dos fuentes de datos independientes: indicadores de movilidad urbana provenientes del TomTom Traffic Index y variables económicas de OECD Cities. En una primera etapa se realizó un proceso de limpieza y estandarización, que incluyó la normalización de nombres de columnas a formato snake_case, la conversión de variables de tiempo a formato datetime, la corrección de tipos numéricos y la estandarización de nombres de ciudades y países para garantizar consistencia entre ambas fuentes.

Dado que los datos de tráfico contienen múltiples registros diarios por ciudad, se llevó a cabo una agregación por ciudad y año, calculando promedios anuales de los principales indicadores de congestión y tiempos de viaje. Este paso permitió transformar la información de alta frecuencia en una unidad de análisis comparable con los datos económicos, los cuales ya se encontraban estructurados a nivel ciudad–año.

Posteriormente, los datasets fueron integrados mediante una unión INNER utilizando las claves city y year, conservando únicamente aquellas ciudades con información disponible tanto de movilidad como de economía. Finalmente, se realizaron validaciones visuales a través de boxplots, histogramas y gráficos comparativos para evaluar distribuciones, detectar valores atípicos (outliers) y explorar tendencias generales entre congestión urbana y productividad económica, asegurando la coherencia y calidad del dataset final antes del análisis interpretativo.

**Hallazgos iniciales:**  

El análisis exploratorio muestra que no existe una relación lineal clara entre los niveles de congestión urbana y el PIB per cápita. Las visualizaciones indican que ciudades con alta congestión vehicular no necesariamente presentan altos niveles de productividad económica, y viceversa. En particular, varias ciudades latinoamericanas exhiben valores elevados de JamsDelay combinados con niveles medios o bajos de PIB per cápita, lo que sugiere que la congestión podría estar actuando como un factor que limita la eficiencia económica urbana.

Asimismo, se identificaron valores atípicos (outliers) tanto en los indicadores de congestión como en el PIB per cápita. El boxplot de JamsDelay revela que un pequeño grupo de ciudades concentra retrasos extremadamente altos, lo que eleva el promedio general y evidencia una distribución altamente asimétrica. De manera similar, el histograma de PIB per cápita muestra una concentración de ciudades en rangos medios y una cola hacia valores altos, indicando desigualdad económica entre ciudades. Estos outliers representan casos relevantes que podrían requerir un análisis más profundo, ya sea para validar la calidad de los datos o para comprender factores estructurales específicos que expliquen su comportamiento extremo.

**Recomendaciones**  
Con base en los hallazgos del análisis, se recomienda priorizar inversiones en infraestructura de transporte en ciudades donde se observa una combinación persistente de alta congestión vehicular y bajos indicadores de productividad económica. En particular, Bogotá y Lima destacan como los casos más críticos dentro del conjunto analizado, ya que ambas presentan altos niveles de congestión (JamsDelay) junto con PIB per cápita significativamente menores en comparación con ciudades como Buenos Aires.

Entre estas, Bogotá muestra la señal más clara de correlación negativa entre movilidad y productividad, al ubicarse consistentemente entre las ciudades con mayor congestión y, al mismo tiempo, registrar niveles de PIB per cápita relativamente bajos.
